```julia
function tobins(i, nbit)
    @assert 1 ≤ i ≤ 2^nbit
    mask = 1 << (nbit-1)
    bin = ones(Int, nbit)
    for n in 1:nbit
        bin[n] = (mask & (i-1)) >> (nbit-n) + 1
        mask = mask >> 1
    end
    return bin
end

function frombins(bin)
    @assert all(1 .≤ bin .≤ 2)
    nbit = length(bin)
    i = 1
    tmp = 2^(nbit-1)
    for n in eachindex(bin)
        i += tmp * (bin[n] -1)
        tmp = tmp >> 1
    end
    return i
end

@assert frombins(tobins(4, 10)) == 4
```

```julia
beta = 1000.0

#ek(kx) = 2*cos(kx)
ek(k) = 2*cos(k) + 2*cos(20k) + cos(5k)

v = FermionicFreq(1)
iv = valueim(v, beta)

gk(kx::Float64) = 1 / (iv - ek(kx))

function gkb(b)
    nbit = length(b)
    N = 2^nbit
    ikx = frombins(b)
    @assert ikx ≤ N
    kx = 2π * (ikx - 1)/N
    return abs.(gk(kx))
end
```

```julia
nbit = 14
N = 2^nbit

gkdata = zeros(ComplexF64, N)
for ikx in 1:N
    bx = tobins(ikx, nbit)
    gkdata[ikx] = gkb(bx)
end
```

```julia
plot(abs.(gkdata))
```

```julia
firstpivot = ones(Int, nbit)
@show firstpivot
qtt, ranks, errors = QuanticsGrids.qtt(
    b -> gkb(b .+ 1),
    2,
    firstpivot;
    cutoff=1e-8, maxiter=200
)
;
```

```julia
sites =  [Index(2, "Qubit,k=$k") for k in 1:nbit]
mps = QuanticsGrids.qtt_to_mps(qtt, sites)
```

```julia
@show 2^(nbit/2)
plot(ITensors.dim.(linkinds(mps)))
```

```julia
gkdata_reconst = vec(Array(reduce(*, mps), reverse(sites)))
```

```julia
@show norm(gkdata_reconst .- gkdata)
p = plot(yscale=:log10, ylims=(1e-3, 1e+3))
plot!(p, abs.(gkdata))
plot!(p, 1e+6 * abs.(gkdata_reconst .- gkdata))
```
